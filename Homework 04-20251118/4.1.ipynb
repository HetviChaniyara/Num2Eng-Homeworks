```python
import scipy.sparse as sp
import numpy as np
```

```python
# Tm 
Tm = sp.diags([-1, 4, -1], [-1, 0, 1], shape=(3,3), format='csr')

# negative identity block
I = sp.diags([-1], [0], shape=(3,3), format='csr')
# zero block
zero = sp.csr_matrix((3,3))

A_sparse = sp.bmat([
    [Tm, I, zero],
    [I, Tm, I],
    [zero, I, Tm]
], format='csr')

# convert to dense array
A = A_sparse.toarray()

print(A)
```

```python
np.linalg.det(A)
```

```python
np.linalg.inv(A)
# divide by h^2
```
